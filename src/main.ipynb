```python
import folium as fl
import pandas as pd
```

```python
data_incidents = pd.read_csv('../data/sites_incidents.csv')
```

```python
data_incidents.head()
```

```python
#setting center for our map
center = [32.887425, -6.914740]

# creating map
map = fl.Map(location = center, zoom_start = 14)
```

```python
# loop to fetch data from dataframe
#  {'beige', 'lightred', 'blue', 'cadetblue', 'darkblue', 'green', 'orange', 'darkgreen', 'lightblue', 'purple', 'lightgray', 'darkpurple', 'pink', 'gray', 'white', 'darkred', 'red', 'black', 'lightgreen'}.

for i, j in data_incidents.iterrows():
	location = [j['site_lat'], j['site_lng']]
	fl.Marker(location,popup = f'Incident Site:{j["site_name"]}',icon=fl.Icon(color='orange', icon='triangle-exclamation', prefix='fa')).add_to(map)
```

```python
data_vehicules = pd.read_csv('../data/vehicules_urgence_localisation.csv')
```

```python
data_vehicules.head()
```

```python
# loop to fetch data from dataframe
for i, j in data_vehicules.iterrows():
	location = [j['vehicule_lat'], j['vehicule_lng']]
	if ('camions_pompiers' in j['nom']):
		fl.Marker(location,popup = f'Vehicule type:{j["nom"]}, capacite : {j[""]}',icon=fl.Icon(color='red', icon='fire-extinguisher', prefix='fa')).add_to(map)
	elif ('station_police' in  j['nom']):
		fl.Marker(location,popup = f'Vehicule type:{j["nom"]}',icon=fl.Icon(color='gray', icon='star', prefix='fa')).add_to(map)
	else:
		fl.Marker(location,popup = f'Vehicule type:{j["nom"]}',icon=fl.Icon(color='lightblue', icon='star-of-life', prefix='fa')).add_to(map)
```
